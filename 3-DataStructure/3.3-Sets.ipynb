Sets 
Sets are a build-in data type in python used to store collections of unique items. They are unordered, meaning that the element do not follow a specific order, and They do not allow duplicate elements. Sets are useful for membership tests. eliminating duplicate entries, and performing mathematical set operations like union,intersection,difference,and symmetric difference.


```python
##create a set 

my_set={1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set=set()
print(type(my_empty_set))
```

```python
my_empty_set=set([1,2,3,4,5,6,1])
print(my_empty_set)
```

```python
## basic sets operations 
## Adding and removing elements 
my_set=set([1,2,3,4,5,6,1])
my_set.add(7)
print(my_set)
```

```python
my_set.remove(3)
my_set
```

```python
my_set.remove(129)
my_set
```

```python
my_set.discard(11) # it doesnot through the error 
my_set
```

```python
## Pop method
remove_element=my_set.pop() # first element will be poped or remove
print(remove_element)
print(my_set)
```

```python
#Clear all the element 
my_set.clear()
my_set
```

```python
##Set memebership test
my_set={1,2,3,4,5}
print(3 in my_set)
print(20 in my_set)
```

```python
##Mathematical operation 

set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

###Unition 
union_set=set1.union(set2)
print(union_set)

##Intersection
##common set from both set

intersection_set=set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2) ## it update the set1 with the intersection elements
print(set1)
```

```python
##Difference 
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}
print(set1.difference(set2))

##difference_update
print(set2.difference_update(set1))
```

```python
## Symentric Difference ## common are remove
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}
set1.symmetric_difference(set2)
```

```python
## Sets Methods 
set1={1,2,3}
set2={3,4,5}
## is subset
print(set1.issubset(set2))
set1.update([1, "naran"])
set1

##issuperset
print(set1.issuperset(set2))
```
