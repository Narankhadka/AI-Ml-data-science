Example 1 . Manage  A to do list 


Create a to Do list to keep track of Tasks 


```python
to_do_list=["Learn python","go gym","eat healthy food","earn money"]

## Adding to taks 

to_do_list.append("meet girlfriend")
to_do_list.append("Go for a run")

##Removing a completed task 

to_do_list.remove("Learn python")

#checking if task is in the list 
if "earn money" in to_do_list:
    print("Dont forgot to earn the money for saving ")

print("To do list remaining")
for task in to_do_list:
    print(f"-{task}")
    
```

Example 2: Organinzing Student Grades


Create a list to store and calculate average grades for students

```python
## Organizing student grades
grades=[90,80,47,99,88]

##adding new grades in list 
grades.append(57)

##Calculating the average grade 
average_grades=sum(grades)/len(grades)
print(f"Average Grade:{average_grades:.2f}")

##Finding the heighest and Lowest Grades
heighest_grades=max(grades)
lowest_grades=min(grades)
print(f"Heighest Grade: {heighest_grades}")
print(f"lowest Grade: {lowest_grades}")
```

Example 3: Managing An Inventory 


USe a list to manage inventory items in a store 

```python
## Managing an inventory 
inventory=["apples","banana","orange","grapes"]

#adding an inventory 
inventory.append("avacado")

#Removing an item that is out of stock 
inventory.remove("banana")

#checking if an item is in stock 

item="orange"
if item in inventory:
    print(f"{item}are in stock.")
else:
    print(f"{item} are out of stock.")

##printing the inventory 
print ("Inventory List")

for item in inventory:
    print(f"-{item}")
```

Example 4: Collecting user feedback 

Use a list to collect and analyze user feedback 


```python
#Collecting user feedback 
feedback=["Great service!!","satisfied","could be better","excellent experience"]

##Adding new feedback 
feedback.append("Not happy ")
feedback.append("poor service ")

#Counting specific feedback 

negetive_feedback=sum(1 for comment in feedback if "bad" in comment.lower() or "not happy" in comment.lower())
print(f"Negetive feedback count: {negetive_feedback}")

#printing all feedback 

print("User feedback:")

for comment in feedback:
    print(f"-{comment}")
```
