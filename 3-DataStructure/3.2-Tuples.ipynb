```python
## Immutable Nature of Tuples
## tuples are immutables , meaning their elements cannot be changed once assigned 

#for list
lst=[1,2,3,4,5]
print(lst)

lst[1]="naran"
print(lst)

tuple=(1,2,3,4,5,6)
print(tuple)
tuple[1]="naran"
print(tuple)

```

```python
numbers=(1,2,3,4,5,6)
```

```python
##Tuple methods 
print(numbers.count(1))
print(numbers.index(3)) # which index 
```

```python
##Packing and unpacking tuple
#packing
packed_tuple=1,"hello",3
print(packed_tuple)
```

```python
#Unpacking a tuple
a,b,c =packed_tuple
print(a)
print(b)
print(c)
```

```python
##Unpacking with star(*)

numbers=(1,2,3,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
##Nested Tuple
#list ho yo
lst=[[1,2,3,4,5],[6,7,8,9],[1,"hello","hi",3.2]]
print(lst[0][::-1])
```

```python
#Nested tuple
nested_tuple=((1,2,3,4,5),(6,7,8,9),(1,"hello","hi",3.2))
print(nested_tuple[0])
print(nested_tuple[1])
print(nested_tuple[2])
print(nested_tuple[0][1])

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()


```

Conclusion
Tuples are versatile and useful in many real-world scenarios where an immutable and ordered collection of items is required. they are commonly used in data structure, function arguments and return values, and as dictionary keys, Understading how to leverage tuples effectively can improve the efficiency and readability of python code
