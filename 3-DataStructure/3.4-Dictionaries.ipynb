Introduction to Dictionaries    


Dictionaries are unordered collections of items. They store data in key-value pairs. Keys must be unique and immutables(e.g, strings, numbers, or tuples), while values can be of any type

```python
## creating dictionaries 
empty_dict={}
print(type(empty_dict))
```

```python
empty_dict=dict()
empty_dict
```

```python
student={"name":"krish","age":20,"grade":24}
print(student)
print(type(student))
```

```python
##Single key is always used 
student={"name":"krish","age":20,"grade":24}
student={"name":"krish","age":20,"grade":24}
print(student)
```

```python
##accessing Dictionary Elements 
student={"name":"krish","age":20,"grade":'A'}
print(student)
```

```python
print(student['grade'])
```

```python
#Accessing using get() methos 
print(student.get('grade',"Not available"))
print(student.get('lastName'))
print(student.get('last_name',"Not available")) ##default message
```

```python
#modifying dictionary Elements 
## Dictionary are mutable, so you can add , update or delete elements 
print(student)
```

```python
student["age"]=33 ## Update value for the key 
print(student)
student["address"]="nepal" ## added a new key and value 
print(student)
```

```python
##Delete the key
del student['grade']
print(student)
```

```python
print(student)
```

```python
##Dictionary methods 

keys=student.keys() ## get all the keys 
print(keys)
values=student.values() ## all the values 
print(values)
```

```python
items=student.items() ## get all key value pairs
print(items)
```

```python
##Shallow copy 
student_copy=student
print(student)
print(student_copy)
```

```python
student["name"]="narankhadka"
print(student)
print(student_copy)
```

```python
student_copy1=student.copy() ## shallow copy ## yasle first ma copy gareko kura lai nai point garirako hunxa change hudaina 

print(student)
print(student_copy1)
student["name"]="sarthak"
print(student_copy1)
print(student)
```

```python
##Iterating Over Dictionaries 
## We can use loops to iterate over dictionaries , key , valuues or items 

## Iterating over keys

for keys in student.keys():
    print(keys)

```

```python
##Ierating over values 
for values in student.values():
    print(values)
```

```python
## Itereaing overs key values pairs 

for key , value in student.items():
    print(f"{key}:{values}")
```

```python
## nested Dictorinaries 
# dict inside another dict 
students={
    "student1":{"name":"naran","age":23},
    "student2":{"name":"sarthak","age":21}
}
print(students)
```

```python
## access nested dictionaries elements 
print(students["student2"]["name"])
print(students["student1"]["name"])
```

```python
students.items()
```

```python
## Ierating Over nested dictonries 
for student_id , student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionary Comprehension 
squares={x:x*2 for x in range(5)}
print(squares)
```

```python
## Condition Dictionary comprehension 
evens={x:x**2 for x in range(10) if x%2==0}
print(evens)

```

```python
## Practical Examples

## Use a dictionary to count the frequency of elements in list 
numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}
for number in numbers:
    if number in frequency: 
       frequency[number]+=1
    else:
        frequency[number]=1

print(frequency)
```

```python
## marge 2 dictionaires into one 

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}

merged_dict={**dict1,**dict2}
print(merged_dict)
```
